```python
import pandas as pd
import os
import numpy as np
```

```python
def loadGRKota():
    folder_path = 'data'
    dfs = []
    for filename in os.listdir(folder_path):
        if filename.startswith('GR Kota'):
            file_path = os.path.join(folder_path, filename)
            df = pd.read_csv(file_path)
            dfs.append(df)
    return dfs

def loadGRProvinsi():
    folder_path = 'data'
    dfs = []
    for filename in os.listdir(folder_path):
        if filename.startswith('GR Provinsi'):
            file_path = os.path.join(folder_path, filename)
            df = pd.read_csv(file_path)
            dfs.append(df)
    return dfs

def loadP0Provinsi():
    folder_path = 'data'
    dfs = []
    for filename in os.listdir(folder_path):
        if filename.startswith('P0 Provinsi'):
            file_path = os.path.join(folder_path, filename)
            df = pd.read_csv(file_path)
            dfs.append(df)
    return dfs
```

```python
dfsGRKota = loadGRKota()
dfsGRProvinsi = loadGRProvinsi()
dfsP0Provinsi = loadP0Provinsi()
```

Cleaning GR Kota

```python
def cleanDfGRKota(df):
    # set second row as header
    df.columns = df.iloc[1]
    # remove first and second row
    df = df[2:]
    # remove column 2016 and 2017 if they exist
    cols_to_drop = []
    for col in df.columns:
        col_str = str(col).strip()
        if '2016' in col_str or '2017' in col_str:
            cols_to_drop.append(col)
    
    if cols_to_drop:
        df = df.drop(columns=cols_to_drop)
    return df
```

Cleaning GR Provinsi

```python
def cleanDfGRProvinsi(df):
    # Drop first three rows
    df = df.iloc[3:].reset_index(drop=True)

    # Set the first row as header
    df.columns = df.iloc[0]
    df = df[1:].reset_index(drop=True)

    # Drop the first column
    df = df.iloc[:, 1:]
    # Replace "-" with NaN
    df = df.replace("-", np.nan)

    # Drop columns and rows where all values are NaN
    df = df.dropna(axis=0, how='all')

    return df
```

```python
def add_val_column(df):
    # Convert all columns to numeric, coercing non-numeric values to NaN
    df = df.apply(pd.to_numeric, errors='coerce')
    
    def compute_val(row):
        # Check last column (already converted to numeric)
        last_val = row.iloc[-1]
        if not pd.isna(last_val):
            return last_val
        
        # Check second and third last columns (already numeric)
        second_last = row.iloc[-2]
        third_last = row.iloc[-3]
        
        valid = []
        for val in [second_last, third_last]:
            if not pd.isna(val):
                valid.append(val)
        
        if len(valid) >= 2:
            return sum(valid) / len(valid)
        elif valid:
            return valid[0]
        else:
            # All three columns are NaN; use row average
            return row.mean(skipna=True)
    
    df['val'] = df.apply(compute_val, axis=1)
    return df
```

```python
dfsGRProvinsi = [cleanDfGRProvinsi(df) for df in dfsGRProvinsi]
dfsGRProvinsi = [add_val_column(df) for df in dfsGRProvinsi]
```

```python
# remove all columns except val
dfsGRProvinsi = [df[['val']] for df in dfsGRProvinsi]
```

```python
def cleanDfP0Provinsi(df):
    # Assume the first column is the region/province name
    keep = []
    keep_flag = False
    for val in df.iloc[:, 0]:
        if isinstance(val, str):
            if val.strip().upper() == "SUMATERA UTARA":
                keep_flag = True
            elif val.strip().upper() == "SUMATERA BARAT":
                keep_flag = False
            # Keep if uppercase or within SUMATERA UTARA section
            keep.append(val.isupper() or keep_flag)
        else:
            keep.append(False)
    df = df[keep]
    return df
```

```python
dfsP0Provinsi = [cleanDfP0Provinsi(df) for df in dfsP0Provinsi]
```
