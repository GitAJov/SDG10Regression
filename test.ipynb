```python
import pandas as pd 
import numpy as np
import os
```

```python
def loadGRKota():
    folder_path = 'data'
    dfs = []
    for filename in os.listdir(folder_path):
        if filename.startswith('GR Kota'):
            file_path = os.path.join(folder_path, filename)
            df = pd.read_csv(file_path)
            dfs.append(df)
    return dfs
```

```python
dfGRKota = loadGRKota()
```

```python
def cleandf(df):
    # Set second row as the header
    df.columns = df.iloc[1]
    
    # #remove 1st and 2nd row
    df = df.iloc[2:]

    # Rename first column to 'Wilayah'
    df = df.rename(columns={df.columns[0]: 'Wilayah'})
        
    return df
```

```python
dfGRKota = [cleandf(df) for df in dfGRKota]
```

```python
merged_df = dfGRKota[0]
for df in dfGRKota[1:]:
    merged_df = pd.merge(merged_df, df, on='Wilayah', how='outer')
```
